Import the required libraries:

```python
import folium
from folium.plugins import MarkerCluster
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
```

Create a DataFrame containing the information about the 10 most found animals in Australia. This includes their names, population, species type, image URLs, and coordinates (latitude and longitude).

```python
# Example data
data = {
    'Animal': ['Kangaroo', 'Koala', 'Emu', 'Platypus', 'Wombat', 'Tasmanian Devil', 'Dingo', 'Echidna', 'Wallaby', 'Kookaburra'],
    'Population': [50000000, 43000, 725000, 30000, 100000, 25000, 20000, 1000000, 10000000, 650000],
    'Species': ['Macropus', 'Phascolarctos cinereus', 'Dromaius novaehollandiae', 'Ornithorhynchus anatinus', 'Vombatus ursinus', 'Sarcophilus harrisii', 'Canis lupus dingo', 'Tachyglossus aculeatus', 'Macropus rufogriseus', 'Dacelo novaeguineae'],
    'Image_URL': [
        'https://upload.wikimedia.org/wikipedia/commons/0/0c/Kangaroo_Australia_01_11_2008_-_retouch.JPG',
        'https://upload.wikimedia.org/wikipedia/commons/4/49/Koala_climbing_tree.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/9/9d/Emu_1_-_Tidbinbilla.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/2/22/Platypus.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/1/18/Vombatus_ursinus_-Maria_Island_National_Park.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/5/5a/Sarcophilus_harrisii_taranna.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/5/5f/Canis_lupus_dingo.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/6/6f/Echidna_-_Tidbinbilla.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/8/8c/Macropus_rufogriseus_rufogriseus_Bruny.jpg',
        'https://upload.wikimedia.org/wikipedia/commons/4/4c/Kookaburra_mh_01.jpg'
    ],
    'Latitude': [-30, -35, -32, -37, -42, -41, -25, -35, -33, -34],
    'Longitude': [135, 149, 148, 145, 147, 145, 135, 149, 150, 150]
}

df = pd.DataFrame(data)
```

Now, create an interactive map using folium and add markers for each animal. When you click on a marker, it will display the animal's image, name, population, and species type.

```python
# Create a base map centered on Australia
m = folium.Map(location=[-25, 135], zoom_start=4)

# Add markers for each animal
for idx, row in df.iterrows():
    # Create a popup with the animal's information
    popup_content = f"""
    <b>{row['Animal']}</b><br>
    <img src="{row['Image_URL']}" width="150"><br>
    Population: {row['Population']}<br>
    Species: {row['Species']}
    """
    popup = folium.Popup(popup_content, max_width=300)

    # Add a marker to the map
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup,
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

# Display the map
m
```
